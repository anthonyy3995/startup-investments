```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
offices_df = pd.read_csv('offices.csv')
offices_df.head()
```

```python
offices_df.description
```

```python
company_city_df = offices_df[["office_id", "city"]]
```

```python
funding_rounds_df = pd.read_csv("funding_rounds.csv")
raised_amount_df = funding_rounds_df[["object_id", "raised_amount_usd"]]
```

```python
total_raised_df = raised_amount_df.groupby(["object_id"]).sum().reset_index()
total_raised_df
```

```python
company_city_df
```

```python
merged_df = company_city_df.merge(total_raised_df, left_on = 'office_id', right_on="object_id", how="inner")
```

```python
city_avg_df = merged_df.groupby(["city"]).mean().reset_index()
```
