```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("archive/offices.csv")
```

```python
df.head()
```

```python
company_city_df = df[["object_id", "city"]]
company_city_df = company_city_df.dropna()
```

```python
fund_rounds_df = pd.read_csv("archive/funding_rounds.csv")
```

```python
fund_rounds_df.head()
```

```python
rasied_amount_df = fund_rounds_df[["object_id", "raised_amount_usd"]]
```

```python
total_raised_df = rasied_amount_df.groupby(["object_id"]).sum().reset_index()
```

```python
print(total_raised_df.shape)
print(company_city_df.shape)
```

```python
merged_df = company_city_df.merge(total_raised_df, left_on="object_id", right_on="object_id", how="inner")
```

```python
city_avg_df = merged_df.groupby(["city"]).mean().reset_index()
```
